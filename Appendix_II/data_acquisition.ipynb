###0.4.1 Other Functions

```python
# initialise virtual browser
def vb_ini():
  options = Options()
  options.add_argument("--headless")
  options.add_argument("--no-sandbox")
  #options.headless = True
  service = "/usr/bin/chromedriver"
  global driver 
  driver = webdriver.Chrome(service, options=options)

# get soup object using html.parser
def getdata(url):
  r = requests.get(url)
  soup = BeautifulSoup(r.content, 'html.parser')
  return soup

# check if a string has number character
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

# check if the number type is float
def is_float(num):
  return type(num) == float
```

###0.4.2 Scraping Functions

```python
# retrieve all property URLs in lifestyle community website
def retrive_page_life():
  vb_ini()
  driver.get(URL_LIFE)
  # wait for 3 seconds to load
  time.sleep(3)
  # drop down list of each area
  area_code = 1
  url_list = []
  for area in LIFE_AREA:
    print(f"Area{area_code}: {area}")
    drop = Select(driver.find_element("xpath", drop_Xpath))
    drop.select_by_visible_text(area)
    count = 0
    while True:  
      try:
        noise = 15 + random.uniform(-5, 5)
        WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, load_Xpath))).click()
        time.sleep(noise)
        count += 1
        print('\r',  end='')
        print(f"load more results button clicked {count}", end = "", flush = True)
      except:
        print("\nNo more load more results button to be clicked")
        break
    page = driver.page_source
    soup = BeautifulSoup(page, 'lxml')
    urls = [div_a['href'] for div_a in soup.body.select(".o-grid .col-span-full .h-full a[href]")]
    url_list.append(urls)
    area_code += 1
    print(f"Successfully clicked {count} times and loaded {len(urls)} properties in {area}!\n")
  return url_list

################################################################
######################## FUNCTION BREAK ########################
################################################################


# retrieve all property URLs in ingenia lifestyle website
def retrive_page_inge():
  url_list = []
  for i in range(len(URL_INGE_AREA)):
    url = URL_INGE + URL_INGE_AREA[i]
    # iterate list of each area
    response = requests.get(url)
    urls = []
    if response.status_code == 200:
      html_content = response.text
      soup = BeautifulSoup(html_content, 'lxml')
      property_elements = soup.find_all("div", class_="template-property cell small-12 medium-6 large-4")
      # loop over each property element
      for property_element in property_elements:
        inner_element = property_element.find("div", class_="inner")
        a_elements = inner_element.find_all("a")
        # loop over each anchor element
        for a_element in a_elements:
          href = a_element['href']
          urls.append(href)
          url_list.append(href)
      print(f"Successfully loaded {len(urls)} properties in {INGE_AREA[i]}!\n")
    else:
      print("Failed to get the HTML content.")
  return url_list

################################################################
######################## FUNCTION BREAK ########################
################################################################


# scrape property data in the url of lifestyle community
def life_data_scraping(link):
  soup = getdata(link)
  # Initialisation
  prop = prop_initial.copy()
  prop_comm = prop_initial["Community Name"]
  prop_title = prop_initial["Home Type"]
  prop_price = prop_initial["Price (Total RRP) $"]
  prop_type = prop_initial["New Home or Resale?"]
  prop_lot = prop_initial["Lot No"]
  prop["Type of car space (garage etc) drop down box, car port or garage"] = "Carport"
  prop["State"] = "VIC"
  prop["Link to advertisement/photos etc"] = link

  # COMMUNITY
  try:
    prop_comm = soup.find_all(class_="u-body-xl text-heading mb-2")[0].text
  except:
    pass
  prop["Community Name"] = prop_comm
  try:
    prop["Community ID"] = COMMUNITIES[prop_comm]
  except:
    prop["Community ID"] = COMMUNITIES["null"]
  
  # NAME
  try:
    prop_title = soup.find_all(class_="u-h2 text-primary-navy")[0].text
  except:
    pass
  prop["Home Type"] = prop_title

  # PRICE, TYPE
  try:  
    prop_price = soup.find(class_="u-h4").text
    prop_price = prop_price.strip()
    if len(prop_price) != 0:
      if "from" in prop_price:
        prop_type = "New Home"
      else:
        prop_type = "Resale"
  except:
    pass
  prop["Price (Total RRP) $"] = re.sub('[^0-9]', '', prop_price)
  prop["New Home or Resale?"] = prop_type
  
  # BUILT, LOT
  if has_numbers(prop_title):
    prop["Established or OTP?"] = "Established"
    if "#" in prop_title:
      prop_lot = re.sub('[^0-9]', '', prop_title)
      prop_title = ''.join(re.findall(r'[a-zA-Z\s]+', prop_title))
    elif "-" in prop_title:
      prop_lot = re.sub('[^0-9]', '', prop_title)
      prop_title = ''.join(re.findall(r'[a-zA-Z\s]+', prop_title))
  prop["Lot No"] = prop_lot
  prop["Home Type"] = prop_title.strip()
  
  # BUILDING TYPE (extracted from home type)
  try:
    prop["Type"] = re.match(r'^(\w+)', prop_title.strip()).group(1)
    if " II" in prop_title:
      prop["Type"] += "2"
  except:
    pass

  # FACADE
  prop_facade = 1
  if prop_type == "New Home":
    try:
      prop_facade = int(soup.find_all(class_="js-gallery")[0]["data-num-images"])
      if prop_facade >= 5:
        prop_facade = 1 # too many images means it's an established home
    except:
      pass
  prop["Number of façade options?  (drop down box and text box titled \"D"] = prop_facade
  
  # ATTRIBUTES
  try: 
    prop_char = soup.find_all(class_="u-body-md ml-4")
    char_list = [prop_char[i].text for i in range(len(prop_char))]
    char_list_refine = [i for i in char_list if not i.isnumeric()]
    for elem in char_list_refine:
        try:
            num, char = elem.split(" x ")
            if char[-1] == "s":
              char = char[:-1]
            prop[char.capitalize()] = int(num)
        except:
            prop[elem.capitalize()] = 1
    if "Carport" in prop:   
      prop["Driveway Parking"] = prop["Carport"]
      prop["Car space no"] = prop.pop("Carport")
    if "Bathroom" in prop and "Toilet" in prop:
      if prop["Toilet"] > prop["Bathroom"]:
        prop["Powder room  (WCs) - Drop down"] = prop["Toilet"] - prop["Bathroom"]
    if "Bathroom" in prop:
      prop["Bathroom no (note WCs) Drop Down"] = prop.pop("Bathroom")
    if "Bedroom" in prop:
      prop["Bedrooms Drop Down"] = prop.pop("Bedroom")
  except:
    pass

  # delete unwanted key
  if "2.5" in prop:
    del prop["2.5"]
  
  return prop

################################################################
######################## FUNCTION BREAK ########################
################################################################


# scrape property data in the url of ingenia lifestyle

# filter page style
def pre_process(link):
  soup = getdata(link)
  div_style = soup.select_one('div.logo')
  style = div_style['style']

  if style == style_1:
    prop = inge_old_data_scraping(link)
  elif style == style_2:
    prop = inge_new_data_scraping(link)
  return prop

################################################################
######################## FUNCTION BREAK ########################
################################################################


# scraping settings suitable for both page style
def inge_master_info(link):
  soup = getdata(link)
  # Initialisation of the property dictionary
  prop = {}
  prop.update(prop_initial)
  prop_comm = prop_initial["Community Name"]
  prop_title = prop_initial["Home Type"]
  prop_price = prop_initial["Price (Total RRP) $"]
  prop_type = prop_initial["New Home or Resale?"]
  prop_lot = prop_initial["Lot No"]
  prop["Type of car space (garage etc) drop down box, car port or garage"] = "Garage"
  prop["Link to advertisement/photos etc"] = link

  # COMMUNITY
  prop_comm = re.findall(r'/([^/]+)/[^/]+$', link)[0]
  prop["Community Name"] = "Ingenia " + prop_comm.replace('-', ' ').title()
  try:
    prop["Community ID"] = COMMUNITIES[prop["Community Name"]]
  except:
    prop["Community ID"] = COMMUNITIES["null"]

  # STATE
  if "nsw" in link:
    prop["State"] = "NSW"
  elif "qld" in link:
    prop["State"] = "QLD"
  elif "vic" in link:
    prop["State"] = "VIC"

  # HOME TYPE, LOT NO
  prop_title = re.findall(r'/([^/]+)$', link)[0]
  prop_title = ' '.join(word.lower() for word in prop_title.split('-')[:-1])
  if re.search(r'^t[\d]+', prop_title):
    prop_title = prop_title[1:]
  if "lot" in prop_title:
    prop_title = prop_title.replace("lot ", "")
  title_leader = prop_title.split(' ')[0]
  if title_leader.isnumeric():
    prop["Lot No"] = title_leader
    prop_street = ' '.join(word.title() for word in prop_title.split(' ')[1:])
  else:
    match = re.match(r'^([\D]+)([\d\w\s]+)$', prop_title)
    if match:
      prop["Home Type"] = match.group(1).title()
      prop_street = match.group(2)
  prop["Type"] = prop["Home Type"].strip()
  
  # STREET NO, STREET ADDRESS
  match = re.match(r'^(\d+)(?:\s+([\d]+))?(\s+.+)$', prop_street)
  if match:
    street_number = match.group(1)
    if match.group(2):
      street_number_range = match.group(2)
      street_number += "-" + street_number_range
    street_address = match.group(3).strip()
  try:
    prop["Street Number"] = street_number
    prop["Street Address"] = street_address.title()
  except:
    pass

  # PRICE, TYPE, BUILT
  prop_price = ""
  prop_type = "Resale"
  key_words = ["enquire", "future", "new", "from", "contact"]
  try:  
    prop_price = soup.find(class_="price").text
    prop_price = prop_price.strip()
    if len(prop_price) == 0:
      prop_type = "New Home"
    else:
      for key_word in key_words:
        if key_word in prop_price.lower():
          prop_type = "New Home"
  except:
    pass
  
  if "-" in prop_price:
      prop_price = prop_price.split("-")[0]
  prop["Price (Total RRP) $"] = re.sub('[^0-9]', '', prop_price)
  prop["New Home or Resale?"] = prop_type
  if prop_price:
    prop["Established or OTP?"] = "Established"

  return prop

################################################################
######################## FUNCTION BREAK ########################
################################################################


# scraping settings suitable for old page style
def inge_old_data_scraping(link):
  soup = getdata(link)
  prop = inge_master_info(link)

  # ATTRIBUTES
  features = soup.find_all("div", class_="feature cell small-4 medium-12")
  for feature in features:
    label = feature.find_all('div', class_="wrapper")[-1].next_sibling.strip()
    if feature.find("strong"):
      value = feature.find('strong').text.strip()
    else:
      value = 1
    if "Carpark" in label:
      prop["Driveway Parking"] = int(value)
      prop["Car space no"] = int(value)
    if "Bathroom" in label:
      if value.isdigit() and "." not in value:
        prop["Bathroom no (note WCs) Drop Down"] = int(value)
      else:
        prop["Bathroom no (note WCs) Drop Down"] = int(float(value))
        prop["Powder room  (WCs) - Drop down"] = 1
    if "Bedroom" in label:
      prop["Bedrooms Drop Down"] = int(value)
    if "Study" in label:
      prop["Study"] = int(value)

  return prop

################################################################
######################## FUNCTION BREAK ########################
################################################################


# scraping settings suitable for new page style
def inge_new_data_scraping(link):
  soup = getdata(link)
  prop = inge_master_info(link)

  # ATTRIBUTES
  try:
    features = soup.find_all("div", class_="feature")
    for feature in features:
      label = feature.find('div', class_='wrapper').text
      if feature.find("strong"):
        value = feature.find('strong').text.strip()
      else:
        value = 1
      if "Carpark" in label:
        prop["Driveway Parking"] = int(value)
        prop["Car space no"] = int(value)
      if "Bathroom" in label:
        if value.isdigit() and "." not in value:
          prop["Bathroom no (note WCs) Drop Down"] = int(value)
        else:
          prop["Bathroom no (note WCs) Drop Down"] = int(float(value))
          prop["Powder room  (WCs) - Drop down"] = 1
      if "Bedroom" in label:
        prop["Bedrooms Drop Down"] = int(value)
      if "Study" in label:
        prop["Study"] = int(value)
  except:
    pass 
  
  return prop

################################################################
######################## FUNCTION BREAK ########################
################################################################

def get_home_type(url):
    """
    Returns the home type for a given property URL.
    """
    if "lifestylecommunities" in url:
        soup = getdata(url)
        try:
            name = soup.find_all(class_="u-h2 text-primary-navy")[0].text.strip()
            if has_numbers(name):
                if "#" in name:
                    name = name[5:]
                elif "-" in name:
                    name = name[:-6]
        except:
            name = "Sorry, cannot find a suitable name"
    elif "ingenialifestyle" in url:
        # HOME TYPE, LOT NO
        name = re.findall(r'/([^/]+)$', url)[0]
        name = ' '.join(word.title() for word in name.split('-')[:-1]).strip()
        title_leader = name.split(' ')[0]
        if title_leader.isnumeric():
            name = "Sorry, cannot find a suitable name"
        else:
            match = re.match(r'^([\D]+)([\d\w\s]+)$', name)
            if match:
                name = match.group(1).title().strip()
    else:
        raise ValueError("The URL is not recognized.\n")

    return name

def find_home_type():
    """
    Prompts the user to enter a property URL and prints the home type.
    """
    while True:
        url = input("Please paste the property URL: ")
        try:
            home_type = get_home_type(url)
            print(f"The home type by the naming convention is: {home_type}.\n")
            while True:
                res = input("Continue to check another URL? (y/n): ").lower()
                if res == "y":
                    print()
                    break
                elif res == "n":
                    print("Thank you, see you.")
                    return
                else:
                    print("Invalid input. Please enter 'y' or 'n'.\n")
        except ValueError as e:
            print(e)
            print()
        except:
            print("Sorry, an error occurred. Please try again.\n")
```
